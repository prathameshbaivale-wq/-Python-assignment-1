```python

# Program for membership operators

my_list = [10, 20, 30, 40, 50]

num = int(input("Enter number to check: "))

if num in my_list:
    print(num, "is present in the list")
else:
    print(num, "is not present in the list")

if num not in my_list:
    print(num, "is not in the list")
```

```python

# Program for indexing and slicing

my_list = [10, 20, 30, 40, 50]

print("List:", my_list)

# Indexing
print("First element:", my_list[0])

# Negative Indexing
print("Last element:", my_list[-1])

# Slicing
print("Elements from index 1 to 3:", my_list[1:4])
```

```python

# Program for list operations

my_list = [10, 20, 30, 40]

# Update
my_list[1] = 25
print("After update:", my_list)

# Append
my_list.append(50)
print("After append:", my_list)

# Insert
my_list.insert(2, 35)
print("After insert:", my_list)

# Delete
my_list.remove(30)
print("After delete:", my_list)
```

```python

# Program for basic list operations

list1 = [1, 2, 3]
list2 = [4, 5, 6]

# Concatenation
print("Concatenation:", list1 + list2)

# Repetition
print("Repetition:", list1 * 2)

# Length
print("Length of list1:", len(list1))

# Maximum
print("Maximum value:", max(list2))

# Minimum
print("Minimum value:", min(list2))
```
