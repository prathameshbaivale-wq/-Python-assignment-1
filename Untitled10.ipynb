```python

import simpy
import random

RANDOM_SEED = 42
NUM_CUSTOMERS = 10
ARRIVAL_INTERVAL = 2
SERVICE_TIME = 3

def customer(env, name, server, wait_times):
    arrival_time = env.now

    with server.request() as request:
        yield request

        wait = env.now - arrival_time
        wait_times.append(wait)
        print(f"{name} waited {wait:.2f} units")

        service_time = random.expovariate(1.0 / SERVICE_TIME)
        yield env.timeout(service_time)

def setup(env, wait_times):
    server = simpy.Resource(env, capacity=1)

    for i in range(NUM_CUSTOMERS):
        yield env.timeout(random.expovariate(1.0 / ARRIVAL_INTERVAL))
        env.process(customer(env, f"Customer {i+1}", server, wait_times))

print("\n--- Single Server Simulation ---")

random.seed(RANDOM_SEED)
env = simpy.Environment()

wait_times = []   # ✅ reset list properly

env.process(setup(env, wait_times))
env.run()

# ✅ Safe average calculation
if len(wait_times) > 0:
    avg_wait = sum(wait_times) / len(wait_times)
else:
    avg_wait = 0

print("Average waiting time:", round(avg_wait, 2))
```

```python
import simpy
import random

wait_times = []

def customer(env, name, server):
    arrival_time = env.now

    with server.request() as request:
        yield request

        wait = env.now - arrival_time
        wait_times.append(wait)
        print(f"{name} waited {wait:.2f}")

        yield env.timeout(random.expovariate(1/3))

def setup(env):
    server = simpy.Resource(env, capacity=2)  # 2 servers

    for i in range(10):
        yield env.timeout(random.expovariate(1/2))
        env.process(customer(env, f"Customer {i+1}", server))

print("\n--- Two Server Simulation ---")
env = simpy.Environment()
env.process(setup(env))
env.run()

print("Average waiting time:", sum(wait_times)/len(wait_times))
```
