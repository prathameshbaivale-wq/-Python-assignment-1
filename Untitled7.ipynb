```python

# Creating a tuple
t = (1, 2, 3, 4)

# Length
print("Length:", len(t))

# Concatenation
t2 = (5, 6)
print("Concatenation:", t + t2)

# Repetition
print("Repetition:", t * 2)

# Membership
print("Is 3 in tuple?", 3 in t)
print("Is 10 in tuple?", 10 in t)
```

```python

t = (10, 20, 30, 40, 50)

# Indexing
print("First element:", t[0])

# Negative Indexing
print("Last element:", t[-1])

# Slicing
print("Slice (1 to 3):", t[1:4])

# Iteration
print("Iterating elements:")
for i in t:
    print(i)
```
