```python

d = {'a': 3, 'b': 1, 'c': 2}

# Ascending
asc = dict(sorted(d.items(), key=lambda x: x[1]))
print("Ascending:", asc)

# Descending
desc = dict(sorted(d.items(), key=lambda x: x[1], reverse=True))
print("Descending:", desc)
```

```python

t = (1, 2, 3)

# Tuples are immutable, so convert to list
temp = list(t)
temp.append(4)

t = tuple(temp)
print("Updated tuple:", t)
```
