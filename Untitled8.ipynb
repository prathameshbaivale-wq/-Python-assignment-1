```python

# Creating a tuple
t = (1, 2, 3, 4)

# Length
print("Length:", len(t))

# Concatenation
t2 = (5, 6)
print("Concatenation:", t + t2)

# Repetition
print("Repetition:", t * 2)

# Membership
print("Is 3 in tuple?", 3 in t)
print("Is 10 in tuple?", 10 in t)
```

```python

t = (10, 20, 30, 40, 50)

# Indexing
print("First element:", t[0])

# Negative Indexing
print("Last element:", t[-1])

# Slicing
print("Slice (1 to 3):", t[1:4])

# Iteration
print("Iterating elements:")
for i in t:
    print(i)
```

```python

t = (1, 2, 3)

# Tuples are immutable
try:
    t[0] = 10
except TypeError:
    print("Tuple is immutable, cannot change value")

# Deleting tuple
del t

# Uncomment below line to check error after deletion
# print(t)
```

```python

# Given sequence (list)
lst = [4, 7, 1, 9, 2]

# Convert list to tuple
t = tuple(lst)

print("Tuple:", t)
print("Length:", len(t))
print("Maximum:", max(t))
print("Minimum:", min(t))
```
