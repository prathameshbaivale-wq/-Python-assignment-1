```python

# Program to display even numbers from 1 to 10

for i in range(1, 11):
    if i % 2 == 0:
        print(i)
```

```python

# Program to add odd numbers from 1 to 10

sum = 0

for i in range(1, 11):
    if i % 2 != 0:
        sum = sum + i

print("Sum of odd numbers =", sum)
```

```python

# Program to print Fibonacci series from 0 to 50

a = 0
b = 1

print("Fibonacci Series:")

while a <= 50:
    print(a)
    c = a + b
    a = b
    b = c
```

```python

# Program to remove characters at odd index positions

text = input("Enter a string: ")

result = ""

for i in range(len(text)):
    if i % 2 == 0:
        result = result + text[i]

print("String after removing odd index characters:", result)
```
